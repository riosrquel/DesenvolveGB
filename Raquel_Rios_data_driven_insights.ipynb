```python
# Configura√ß√µes de estilo
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
#Gr√°fico de Linha ‚Äì Varia√ß√£o de vendas ao longo do tempo
# Converter ORDERDATE em datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Agrupar por data e somar as vendas
vendas_diarias = df.groupby('ORDERDATE')['SALES'].sum().sort_index()

# Plot
vendas_diarias.plot(kind='line', title='Vendas ao Longo do Tempo')
plt.xlabel('Data do Pedido')
plt.ylabel('Total de Vendas')
plt.show()
```

Gr√°fico de Linha: Este gr√°fico mostra a tend√™ncia das vendas ao longo do tempo. √â poss√≠vel identificar per√≠odos de pico ou queda nas vendas, o que pode indicar sazonalidade ou campanhas promocionais

```python
#Gr√°fico de Barras ‚Äì Vendas por Categoria de Produto
vendas_por_categoria = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values()

# Plot
vendas_por_categoria.plot(kind='barh', title='Vendas por Categoria de Produto', color='skyblue')
plt.xlabel('Total de Vendas')
plt.ylabel('Categoria')
plt.show()
```

Gr√°fico de Barras: Este gr√°fico compara o total de vendas entre diferentes categorias de produtos. As categorias com maior faturamento aparecem no topo, ajudando a identificar as mais lucrativas.

```python
#Gr√°fico de Dispers√£o ‚Äì Pre√ßo Unit√°rio vs. Quantidade Vendida
sns.scatterplot(data=df, x='PRICEEACH', y='QUANTITYORDERED')
plt.title('Rela√ß√£o entre Pre√ßo Unit√°rio e Quantidade Vendida')
plt.xlabel('Pre√ßo Unit√°rio')
plt.ylabel('Quantidade Vendida')
plt.show()
```

Gr√°fico de Dispers√£o: Este gr√°fico mostra como a quantidade vendida varia em fun√ß√£o do pre√ßo de cada item. Pode ajudar a entender se h√° rela√ß√£o entre pre√ßo e volume de vendas (por exemplo: itens mais baratos sendo vendidos em maiores quantidades).
